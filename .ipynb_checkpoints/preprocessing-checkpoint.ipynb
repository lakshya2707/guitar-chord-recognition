```python
import os
import json
```

```python
DATASET_PATH = "chords"
JSON_PATH = "my_data2.json"
```

```python
def save_chord_profile(dataset_path, output_json_path):

    # dictionary to store mapping, labels, and pitch classes
    chord_data = {
        "mapping": [],
        "labels": [],
        "pitch": []
    }

    # loop through all chord sub-folder
    for index, (dir_path, dir_names, file_names) in enumerate(os.walk(dataset_path)):

        # ensure we're processing a sub-folder level
        if dir_path is not dataset_path:

            # save chord label (i.e., sub-folder name) in the mapping
            semantic = dir_path.split("\\")[-1]
            chord_data["mapping"].append(semantic)
            print("\nProcessing: {}".format(semantic))

            # process all audio files in chord sub-dir
            for file_name in file_names:

                # load audio file
                file_path = os.path.join(dir_path, file_name)

                # process all segments of audio file
                ptc = PitchClassProfiler(file_path)
                chord_data["pitch"].append(ptc.get_profile())
                chord_data["labels"].append(index - 1)
                print("{}, segment:{}".format(file_path, 1))

    # save pitch classes to json file
    with open(output_json_path, "w") as json_file:
        json.dump(chord_data, json_file, indent=4)
```

```python
if __name__ == "__main__":
    save_chord_profile(DATASET_PATH, JSON_PATH)
```

```python
import json
import matplotlib.pyplot as plt

# Load data from JSON file
with open('my_data2.json', 'r') as file:
    data = json.load(file)

# Extract pitch values
pitch_values = data['pitch']

# Define pitch categories
pitch_categories = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']

# Initialize dictionary to store counts
pitch_count = {pitch: 0 for pitch in pitch_categories}
```

```python
pitch_count
```

```python
# Count occurrences of each pitch
%matplotlib inline
count=0
mapping=["a","am","bm","c","d","dm","e","em","f","g"]
for pitches in pitch_values:
    count=count+1
    for i, pitch_value in enumerate(pitches):
        pitch_name = pitch_categories[i]
        pitch_count[pitch_name] += pitch_value
    if(count%200==0):
        plt.figure(figsize=(6,4))
        normalized_values = [value / 200 for value in pitch_count.values()]
        plt.bar(pitch_count.keys(), normalized_values, color='skyblue')
        plt.title(f'Real pcp representation of chord {mapping[(count-1)//200]}')
        plt.xlabel('Pitch')
        plt.ylabel('average_value')
        plt.xticks(rotation=45)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
        pitch_count = {key: 0 for key in pitch_count}
```
